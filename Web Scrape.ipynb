```python
def find_header(url):
    page= requests.get(url)
    if page.status_code == 200:
        data= BeautifulSoup(page.content, 'html.parser')
        header  = data.find_all('span',class_='mw-headline')
        header_e=[]
        for i in header:
            header_e.append(i.text.replace('\n',''))
        header_e
        return(header_e)
    else :
        return ("Permission denied")
```

```python
def Odi_team_men(url):
    matc=[]
    points=[]
    ratingss=[]
    
    page4= requests.get(url)
    data4 = BeautifulSoup(page4.content, 'html.parser')
    team=data4.find_all(class_='u-hide-phablet')
    team_name=[]
    for i in range(0,10):
        team_name.append(team[i].text)
    print("Top 10 Odi team in Men ->")
    print(team_name)
    print("---------------------------------------------")
    
    matches_top=data4.find_all(class_='rankings-block__banner--matches')
    matc.append(matches_top[0].text)
    #print(matc)
    
    points_top=data4.find(class_="rankings-block__banner--points")
    points.append(points_top.text)
    #print(points_top.text)
    
    ratings_top=data4.find_all(class_='rankings-block__banner--rating u-text-right')
    ratingss.append(ratings_top[0].text.strip())
    #print(ratings_top)
    
    match=data4.find_all(class_="table-body__cell u-center-text")
    match
    
    for i in range(0,18):
        if i %2 != 0:
            points.append(match[i].text)
        else:
            matc.append(match[i].text)
    print("Top 10 Odi team Match in Men ->")
    print(matc)
    print("---------------------------------------------")
    print("Top 10 Odi team Points in Men ->")
    print(points)
    print("---------------------------------------------")
    ratings=data4.find_all(class_='table-body__cell u-text-right rating')
    
    for i in range (0,9):
        ratingss.append(ratings[i].text)
    print("Top 10 Odi team Ratings in Men")
    print(ratingss)
```

## Task 8

```python
def forecast(url):
    periodss=[]
    shrt_descs=[]
    temps=[]
    temp_high_=[]
    temp_low_=[]
    detailed_forecast_periods=[]
    detailed_forecast_descrips=[]
    
    page8=requests.get(url)
    data8 = BeautifulSoup(page8.content, 'html.parser')
    
    period=data8.find_all(class_="period-name")
    for i in range(0,9):
        periodss.append(period[i].text)
    #periodss
    
    shrt_desc=data8.find_all(class_="short-desc")
    for i in range(0,9):
        shrt_descs.append(shrt_desc[i].text)
    #shrt_descs
    
    temp_low=data8.find_all(class_="temp temp-low")
    #print(temp_low)
    temp_high=data8.find_all(class_="temp temp-high")
    #temp_high
    
    for i in range(0,4):
        temp_low_.append(temp_low[i].text)
    #temp_low_
    
    for i in range(0,5):
        temp_high_.append(temp_high[i].text)
    #temp_high_
    temps = [None]*(len(temp_high_)+len(temp_low_))
    temps[::2] = temp_high_
    temps[1::2] = temp_low_
    #temps
    
    detailed_forecast_period=data8.find_all(class_="col-sm-2 forecast-label")
    for i in range (len(detailed_forecast_period)):
        detailed_forecast_periods.append(detailed_forecast_period[i].text)
    #detailed_forecast_periods
    
    
    detailed_forecast_descrip=data8.find_all(class_="col-sm-10 forecast-text")
    #detailed_forecast_descrip
    for i in range (len(detailed_forecast_descrip)):
        detailed_forecast_descrips.append(detailed_forecast_descrip[i].text)
        
    #detailed_forecast_descrips
    
    print("Periods on the page are as -> \n")
    print(periodss)
    print("-----------------------------------------------------\n")
    print("Short Descriptions are as -> \n")
    print(shrt_descs)
    print("-----------------------------------------------------\n")
    print("Temperatures are as -> \n")
    print(temps)
    print("-----------------------------------------------------\n")
    print("Detailed Forecasts are as -> \n")
    #print(detailed_forecast_descrips)
    for i in range(len(detailed_forecast_periods)):
        print(detailed_forecast_periods[i])
        print(detailed_forecast_descrips[i])
        print("-----------------------------------------\n")
    
```
